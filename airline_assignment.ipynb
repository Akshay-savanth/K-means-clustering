```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
airline=pd.read_excel("EastWestAirlines.xlsx",sheet_name='data')
airline
```

```python
airline=airline.drop(['ID#'],axis=1)
airline
```

```python
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(airline.iloc[:,1:]))
data 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
airline_subset = pd.DataFrame(scaler.fit_transform(airline.iloc[:,1:]))
airline_subset 
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
z = linkage(data, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram( z,)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
h_complete = AgglomerativeClustering(n_clusters=5, linkage='complete',affinity = "euclidean").fit(data) 

cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
airline['clust']=cluster_labels # creating a  new column and assigning it to new column 
airline
```

```python
airline.iloc[:,1:].groupby(airline.clust).mean()
```

```python
# K-MEANS CLUSTERING METHOD
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 7):
    clf = KMeans(n_clusters=i)
    clf.fit(data)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 7), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(data)
```

```python
y_kmeans
#clf.cluster_centers_
clf.labels_
```

```python
ab=pd.Series(y_kmeans)  # converting numpy array into pandas series object 
airline['clust']=ab# creating a  new column and assigning it to new column 
airline
```

```python
airline.iloc[:,1:7].groupby(airline.clust).mean()
```

```python
airline.plot(x="Balance",y ="Bonus_miles",c=clf.labels_,kind="scatter",s=50 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```
